```python
def Sortme(arry, Sortlist):
  while arry:  #unsorted list
    minimum = arry[0]
    for item in arry:
        if item < minimum:
            minimum = item
    Sortlist.append(minimum)
    arry.remove(minimum)
  return Sortlist

  
```

```python
Sortlist_1 = Sortme(obser_1, Sortlist_1)
Sortlist_1
```

```python
Sortlist_2 = Sortme(obser_2, Sortlist_2)
Sortlist_2
```

```python
def List_Ranking(A):
 
    # creating a Rank Vector
    R = [0 for x in range(len(A))]
 
    # Sweep through all elements
    # in A for each element count
    # the number of less than and
    # equal elements separately
    # in r and s.
    for i in range(len(A)):
        (r, s) = (1, 1)
        for j in range(len(A)):
            if j != i and A[j] < A[i]:
                r += 1
            if j != i and A[j] == A[i]:
                s += 1      
        
        # Use formula to obtain rank
        R[i] = r + (s - 1) / 2
 
    # Return Rank Vector
    return R
 
```

```python
if __name__ == "__main__":
    Sort_1 = Sortlist_1
 
    print("Sorted List of First observations:" + "\n", Sort_1)
    RankedList_1 = List_Ranking(Sort_1)
    print("Ranking of first observation list:"+ "\n", RankedList_1)
    Sort_2 = Sortlist_2
    print("Sorted Second of First observations:" + "\n", Sort_2)
    RankedList_2 =List_Ranking(Sort_2)
    print("Ranking of second observation list:"+ "\n", RankedList_2)
```

```python
# Now, finding D = difference between the two ranks of observation

D = []

zip_object = zip(RankedList_1, RankedList_2)

for ranked_list1_i, ranked_list2_i in zip_object:
    D.append(ranked_list1_i-ranked_list2_i)

D
```

```python
squared_number = [D ** 2 for D in D]

D_square =squared_number
D_square
```

```python
#finding sum of squared values
Dsquare_sum = 0;
for i in D_square:
  Dsquare_sum = Dsquare_sum +i;
print(Dsquare_sum)
```

```python
def pearson_corelation(no_of_obs, sum_of_dff_square):
  n_cube = no_of_obs*no_of_obs*no_of_obs;

  pearson_cor = 0;
  divident = ((sum_of_dff_square) / (n_cube -no_of_obs))
  divident = 6 * divident
  pearson_cor = 1 + divident

  return pearson_cor
```
