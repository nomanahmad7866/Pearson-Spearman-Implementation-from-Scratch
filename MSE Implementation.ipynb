MSE:

MSE is used to find the average squared difference between the predicted  and actual values.

Formula,

M.S.E = 1 /n ( ΣYi - y^)^2, i from 1 to n,
Here,
N = No of Samples

Yi = Actual Target values
y^ = Predicted Values,

To find predicted values (y^), i will use simple linear regression formula which is,

y^ = β0 + β1*xi

where,
β1 = Σ(xi-x^)(yi-y^) / (Σ(xi-x^)^2

β0 is constant and β1 is a regression coefficient.
